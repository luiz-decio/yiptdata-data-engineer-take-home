```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/cleaned/20240724134814_cleaned_movie_data.csv", sep=";")
                 
df.head()
```

```python
# Average Running Time of Movies by Year
plt.figure(figsize=(15, 5))
average_running_time_by_year = df.groupby('year')['running_time'].mean()
average_running_time_by_year.plot(kind='line', marker='o', color='orange')
plt.title('Average Running Time of Movies by Year')
plt.xlabel('Year')
plt.ylabel('Average Running Time (Minutes)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
